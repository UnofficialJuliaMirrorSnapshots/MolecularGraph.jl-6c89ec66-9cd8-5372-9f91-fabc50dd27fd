# Access to public chemical databases

- `MolecularGraphIO` have some convenient methods to retrieve chemical structure data from public databases.

```julia
import Pkg
Pkg.activate("..")

using MolecularGraph
```

## PubChem API

- PubChem provides database access via HTTP API that enables us to search chemicals by their name, properties, structures and etc.
- `MolecularGraphIO.pubchemsdf` searches compound by PubChem Compound ID and stores SDFile into temporary cache store.
- Constant `pubchem_dir` indicates the path of the temporary cache store 

```julia
pubchemsdf("6437877", "Cefditoren Pivoxil")
nothing
```

```julia
mol = sdftomol(open(joinpath(pubchem_dir, "Cefditoren Pivoxil.mol")))
mol_svg = drawsvg!(mol, 300, 300)
display("image/svg+xml",  mol_svg)
```
